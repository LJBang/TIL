## I. Matplotlib 시작하기

- 파이썬의 데이터 시각화 라이브러리  
- %matplotlib inline을 통해서 활성화


라이브러리 vs 프레임워크  
- 라이브러리는 내부의 코드를 조합해서 개발
- 프레임워크는 틀이 있어서 틀을 채워서 개발

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## II. Matplotlib Case Study

```python
plt.plot([1, 2, 4, 3, 5])
plt.show()
```

```python
# figsize: figure의 크기 조절
plt.figure(figsize=(3,3)) # plotting을 할 도면 선언
plt.plot([0, 1, 2, 3, 4])
plt.show()
```

### 2차함수 그래프 w. plot()

```python
x = np.arange(-10, 10, 0.01)
y = x**2

# 축 설명
plt.xlabel("x value")
plt.ylabel('f(x) value')

# x, y 범위 설정
plt.axis([-5, 5, 0, 25])

# x, y 눈금 설정
plt.xticks([i for i in range(-5, 6, 1)])
plt.yticks([i for i in range(0, 27, 3)])

# 그래프 타이틀
plt.title("y = x^2 graph")

plt.plot(x, y, label="trend")

#범례, 범례는 plot뒤에 작성해야함
plt.legend()

plt.show()
```

## III. Matplotlib Case Study

```python
# 꺾은선 그래프
x = np.arange(20)
y = np.random.randint(0,20,20)

plt.axis([0,20,0,20])

plt.yticks([i for i in range(0, 21, 5)])

plt.plot(x, y)
plt.show()
```

```python
# 산점도

plt.scatter(x, y)
plt.show()
```

```python
# 박스플롯 
# 주로 수치형 데이터에 대한 정보 (사분위수)

plt.boxplot(y)
plt.title("y value")
plt.show()
plt.boxplot((x,y))
plt.title("x, y value")
plt.show()
```

```python
# 막대그래프
# 범주형데이터에 자주 사용

plt.bar(x, y)
plt.xticks(np.arange(0,20,1))
plt.show()
```

```python
# 히스토그램
# 도수분포를 막대형태로
# 0,1,2 가 아니라 0~2의 범주형 데이터로 구성하고 그림을 그림

plt.hist(y, bins=np.arange(0,22,2))
plt.xticks(np.arange(0,20,2))
plt.show()
```

```python
# 파이차트
# 비율 확인에 용이
z = [100,300,200,400]
plt.pie(z, labels=['1','2','3','4'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

```python
import seaborn as sns

# 커널밀도 - 연속적인 분포를 곡선화
x = np.arange(0,22,2)
y = np.random.randint(0,20,20)

plt.hist(y, bins=x)
plt.show()
sns.kdeplot(y, shade=True)
plt.axis([0,20,0,0.1])
plt.show()
```

```python
# 카운트 그림
# 범주형 컬럼의 빈도를 시각화

vote_df = pd.DataFrame({'name':['Andy', 'Bob', 'Cat'], 'vote':[True, True, False]})
print(vote_df)

# groupby 이용
vote_count = vote_df.groupby(vote_df['vote']).count()
print(vote_count)

plt.bar(x=[False, True], height=vote_count['name'])
plt.show()
```

```python
# sns.countplot 이용
sns.countplot(x = vote_df['vote'])
plt.show()
```

```python
# 캣플롯
# 숫자형 변수와 범주형 변수의 관계를 보여줌

covid = pd.read_csv('./data/country_wise_latest.csv')
s = sns.catplot(x='WHO Region', y='Confirmed', data=covid)
s.fig.set_size_inches(10,6)
plt.show()
```

```python
# 스트립플롯, 스웜플롯
# 점으로 데이터 표현
sns.stripplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
sns.swarmplot(x='WHO Region', y='Recovered', data=covid)
plt.show()
```

```python
# 히트맵
# 행렬을 색상으로

c = covid.corr()
sns.heatmap(c)
plt.show()
```
