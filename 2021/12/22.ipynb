### Pandas groupby
- Spilt : 특정한 기준을 바탕으로 DF분할
- Apply : (통계)함수 `sum(), mean(), median() 등`을 적용해서 각 데이터 압축
- Combine : Apply 결과를 바탕으로 새로운 Series 생성(group key : applied_value)

```python
# WHO Region별 확진자
# split만 적용
cvd_by_region = df['Confirmed'].groupby(by=df['WHO Region'])
# apply and combine
cvd_by_region.sum()
```

```python
# 국가당 감염자 수
cvd_by_region.mean()
```
