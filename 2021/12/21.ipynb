### 통계적 가설검정
표본평균이 알려진 값보다 얼마나 커야 모평균이 그 값보다 크다고 할 수 있을것인가  
가설을 세우고 이를 검증하는 것.  

귀무가설: $H_0 : \mu = \mu_0$  
대립가설: $H_1 : \mu > \mu_0$  
귀무가설이 참이라고 가정할 때, 랜덤하게 선택한 표본에서 표본평균이 나올 확률  
이게 낮으면 귀무가설을 기각  

가설검정  
확률이 낮다는 기준점이 필요 -> 유의수준 $\alpha$ 필요, 알파는 적절한 값 설정  
검정을 위해서 만든 값을 검정 통계량이라고 함.  
표본평규를 표준정규확률분포 Z로 변환 후 검토  

검정의 단계  
- $H_0, H_1$설정  
- 유의수준 $\alpha$설정  
- 검정 통계량 계산  
- 기각역 또는 임계값 계산  
- 주어진 데이터로부터 유의성 판정  


```
어떤 농장에서 자신들이 생산하는 계란의 평균 무게가 10.5그램이라고 홍보하고 있다.
이에 생산된 계란 30개의 표본을 뽑았더니 다음과 같다.  
[10.7, 11.7, ~~~, 9.2] 
이농장의 홍보가 맞는지 유의수준 5%로 검정하시오.  
```  

- $H_0 = 10.5$
- $H_1 != 10.5$
- $\alpha = 0.05$
- $\bar{X} = 10.43$
- s = 1.11
- Z = $\frac{\bar{X}-\mu}{s/\sqrt{n}}$ = -0.351
- $z_\frac{\alpha}{2} = 1.96 > |-0.351| = |Z|$
- 귀무가설을 기각할 수 없다.  


## 교차 엔트로피
- 자기정보
    - A: 사건
    - $i(A) = log_b\frac{1}{P(A)} = -log_bP(A)$
    - 확률이 높은사건은 정보가 많지 않다.
- 엔트로피 
    - 자기정보의 평균
        - 자기정보*확률의 합
    - 특성
        - $0 <= H(X) <= log_2K$
        - K는 사건의 수
    - 데이터를 표현하는데 필요한 평균비트수를 표현
    - 데이터 압축에 사용 가능

### 교차엔트로피
두 확률분포 P, Q에대해서 교차엔트로피 H(P,Q)는  
확률분포 Q에서 사건 Aj의 자기정보 i(Aj)와 확률 P를 곱한다.  
$H(P,Q) = \sum_j P(A_j)i(A_j) = -\sum_j P(A_j)log_2 Q(A_j)$  
위 값은 P와 Q가 얼마나 유사한지를 표현  
둘이 같으면 H(P,Q) = H(P)  
둘이 다르면 교차엔트로피가 더 크다.  

제곱합보다는 교차엔트로피가 학습속도가 빠르기 때문에,  
분류문제에서 주로 교차엔트로피를 사용한다.  
정답이 [1,0]이고, 모델의 예측값이 [a,b]일때, 정답에 대한 예측값의 교차엔트로피를 계산할 수 있고,  
이 교차엔트로피를 줄이는 방향으로 학습을 진행할 수 있다.
