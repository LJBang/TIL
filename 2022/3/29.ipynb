## 데브코스

### Transformer
각각의 단어의 벡터와 어텐션 가중치를 구했을 때, 새로운 단어의 임베딩 값은 각 단어의 임베딩 값과 그 가중치를 곱한것의 합과 같다. -> 단어간의 관계를 파악할 수 있다. 양방향 관계이다.  

#### 추론단계
각 단어의 임베딩에 대해서 세 개의 가중치를 구할 것.  
각각 쿼리, 키, 밸류 벡터임.  
단어1에 대한 단어j의 영향을 확인하기 위해서는 q1에 kj를 내적한 값을 통해 확인할 수 있다.  
이 값에 소프트맥스함수를 적용한 것을 밸류벡터에 곱하면 zj를 구할 수 있다.
모든 다른 j에 대해서 zj를 더해주면 해당하는 단어의 의미 추론이 가능하다.
