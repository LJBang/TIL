협업필터링이 한단계 발전함 SVD++을 포함한 여러 모델의 앙상블  

### 인기도 기반 추천 유닛 개발
- Cold Start이슈가 존재하지 않음
- 인기도의 기준은? 평점? 매출? 판매량?
- 사용자 정보에 따라 확장 가능 -> 지역기반, 나이기반 인기 아이템 추천
- 완전히 개인화되어있지는 않다
- 특정 카테고리별로 추천하는 것이 유리하기 때문에 분류 체계를 갖는 것이 좋다.  

### 유사도 측정
아이템간 유사도를 구할 수 있으면 평점 정보와 상관없이 추천이 가능 -> 개인화가 안되어있다!  

#### 코사인 유사도
- N차원 공간의 두 벡터간의 각도를 보고 유사도를 판단  
#### 피어슨 유사도
- 코사인 유사도는 크기를 볼 수 없기 때문에 피어슨 유사도를 볼 수도 있다.  
- 각 벡터의 요소에서 평균값을 빼서 보정한 후에 코사인 유사도 계산  
- 모든 벡터가 원점을 중심으로 이동되고 벡터간 비교가 더 쉬워진다.
